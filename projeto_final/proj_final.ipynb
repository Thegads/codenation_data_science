```python
import pandas as pd
pd.set_option('display.max_columns', 181)
```

```python
df = pd.read_csv('estaticos_market.csv')
df.drop(columns = ['Unnamed: 0'], inplace = True)
df.shape
```

```python
df_nan = pd.DataFrame({'Column_Type' : df.dtypes,
                       'NaN_Perc' : df.isna().mean() * 100})
df_nan
```

```python
df_nan.groupby(['Column_Type']).count()
```

```python
mask_cols = list(df_nan.index.values)
```

```python
market = df[mask_cols]
market
```

```python
del df
```

```python
for col in market.columns:
    if col.startswith('fl_'):
        market[col] = market[col].astype(bool)
```

```python
market.dtypes.value_counts()
```

```python
# Separar por tipos de colunas para podermos fazer os tratamentos corretos
numeric = []
categorical = []
boolean = []
numericas = set('if')
for column in mask_cols:
    if market[column].dtype.kind in numericas:
        numeric.append(column)
    elif market[column].dtype.kind == 'b':
        boolean.append(column)
    else:
        categorical.append(column)
print('Numericos:', len(numeric))
print('Booleano:', len(boolean))
print('Categorico:', len(categorical))
```

```python
#Vamos analisar as colunas categoricas
del categorical[0]
del categorical[6]
for col in categorical:
    print(col)
    print(market[col].unique(), '\n')
```

```python
#depois tirar as seguintes colunas
[['nm_divisao'],['dt_situacao']]
```

```python
#preencher as colunas numéricas
mean = market[numeric].mean()  
std  = market[numeric].std()  
market.update(((market[numeric]-mean)/std).fillna(value=0))
market[numeric].count()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
market[categorical].dtypes
```

```python
market_dum = pd.get_dummies(market, columns = categorical)
```

```python
market_dum.head()
```

```python
market_dum = pd.get_dummies(market_dum, columns = boolean)
```

```python
market_dum.head()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
ident = market_dum['id']
market_dum.drop(columns = ['id', 'dt_situacao'], inplace = True)
```

```python
%%time

qtd_neighbors = 5
nearest = NearestNeighbors(n_neighbors=qtd_neighbors, metric = 'cosine')

nearest.fit(market_dum)
```

```python
market_base = pd.concat([ident, market_dum], axis= 1)
```

```python
market_base.shape
```

```python
port1 =  pd.read_csv('estaticos_portfolio1.csv', usecols = ['id'])
port1.set_index('id',inplace=True)

port1_enc = market_base.merge(port1, how = "left", left_index=True,right_index=True)
```

```python
ident1 = port1_enc['id']
port1_enc.drop(columns = ['id'], inplace = True)
```

```python
%%time
neighbors_list = {}

for row in range(port1_enc.shape[0]):
    neighbors_list[row] = nearest.kneighbors(port1_enc.iloc[[row]].values)
```
